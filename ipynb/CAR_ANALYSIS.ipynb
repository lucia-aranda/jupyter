P. valor (pearson). Investigar que es p valor, valores o escalas y significado de acuerdo al porcentaje de correlación.

- **¿Qué es el valor p?**
  El valor p es la **probabilidad** de haber obtenido un estadístico de prueba suponiendo que la **hipótesis nula** es cierta. En otras palabras, es un valor entre 0 y 1 que se utiliza en los contrastes de hipótesis para **rechazar o aceptar** la hipótesis nula.

- **Interpretación del valor p:**
  - Si el valor p es **menor** que el **nivel de significación** (generalmente establecido en 0,05 o 0,01), se rechaza la hipótesis nula (se acepta la hipótesis alternativa).
  - Si el valor p es **mayor** que el nivel de significación, se rechaza la hipótesis alternativa (se acepta la hipótesis nula).

- **Significado:**
  El valor p nos ayuda a diferenciar resultados debidos a la casualidad de aquellos que son **estadísticamente significativos**. Sin embargo, ten en cuenta que no garantiza que una hipótesis sea cierta; simplemente proporciona evidencia estadística para rechazarla o no.
